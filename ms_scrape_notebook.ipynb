```python
# game_record = {"date": "date_string", "country": "co_string", "rank": r_number, "game": "game_string"}
# Define list variables to store scraped data
game_data = []
date_string =[]
co_string = []
r_number = []
game_string = []
image_links =[]
# need to convert timestamp into string for mongodb storage
date_stamp = str(dt.date.today())
```

```python
print(date_stamp)
type(date_stamp)
```

```python
# scrapping loop, two soups, one for game title and rank, the other for the image data
for x, y in url_lists.items():
    url = y
    print(x, url)
    
    browser.visit(y)
    html = browser.html
    soup = BeautifulSoup(html)
    data = soup.findAll('h3', attrs={'class': 'c-subheading-6'})
    image_data = soup.findAll('source')
    for z in range(len(data)):
        date_string.append(date_stamp)
        co_string.append(x)
        r_number.append(z+1)
        game_string.append(data[z].string)
        
    try:
        for image in image_data:
            image_links.append(image['data-srcset'])
    except:
        print("something missing")
       
        
```

```python
# checking data values: print (date_string, co_string, r_number, game_string, image_links)
game_info = [date_string, co_string, r_number, game_string, image_links]
print(game_info)
```

```python
# getting data together, zip makes a list of tuples, have to convert to a list
game_data = zip(date_string, co_string, r_number, game_string, image_links)
# print(list(game_data))
list_game_tuples = list(game_data)
print(list_game_tuples)
```

```python
# load data into a pandas dataframe with colunmns
df = DataFrame(list_game_tuples, columns=['Date', 'Country', 'Rank', 'Game', 'Image'])
print(df)
```
